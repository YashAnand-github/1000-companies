# 1000 companies dataset

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn import preprocessing as pre
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as lr
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
%matplotlib inline
sb.set()

df=pd.read_csv('1000_Companies.csv')
df
```

# Scaling

```python
X=df.iloc[:, :-2]
y=df.iloc[:, -1]

#using RobustScaler for scaling
robust = pre.RobustScaler(quantile_range = (25,75))  #creating obj.
scaled=robust.fit_transform(X[['R&D Spend', 'Marketing Spend', 'Administration']])
X=pd.DataFrame(scaled, columns=['R&D Spend', 'Marketing Spend', 'Administration'])

y=pd.DataFrame(y, columns=['Profit'])
X
#Scaling using MinMaxScaler range=[0, 1]
#scaler=pre.MinMaxScaler().fit_transform(X)
#data=pd.DataFrame(scaler, columns=data.columns)
```

```python
#splitting dataset for modeling

X_train, X_rem, y_train, y_rem = train_test_split(X, y, train_size=0.5, random_state=42)
X_cv, X_test, y_cv, y_test=train_test_split(X_rem, y_rem, test_size=0.5, random_state=42)
print('training data shape- ', X_train.shape)
print('Cross validation data shape- ', X_cv.shape)
print('Testing data shape- ', X_test.shape)
```

```python
lin_reg=lr()
lin_reg.fit(X_train, y_train)

y_train_pred=lin_reg.predict(X_train)
y_cv_pred=lin_reg.predict(X_cv)
y_test_pred=lin_reg.predict(X_test )
print('mse for training set=', mean_squared_error(y_train, y_train_pred))
print('% mas for training set=', mean_absolute_percentage_error(y_train, y_train_pred), end='\n\n')
print('mse for validation set=', mean_squared_error(y_cv, y_cv_pred))
print('% mas for validation set=', mean_absolute_percentage_error(y_cv, y_cv_pred), end='\n\n')
print('mse for test set=', mean_squared_error(y_test, y_test_pred))
print('% mas for test set=', mean_absolute_percentage_error(y_test, y_test_pred), end='\n\n')
print('train score=', lin_reg.score(X_train, y_train).round(2), 'test score=', lin_reg.score(X_test, y_test ).round(2), 
      'validation score=', lin_reg.score(X_cv, y_cv).round(2))
print(lin_reg.coef_, lin_reg.intercept_ )
```
